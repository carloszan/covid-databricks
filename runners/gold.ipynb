```python
from covid.common import SparkSessionService
```

```python
workspace_path = "COVID-DATABRICKS"
database_path = f"/{workspace_path}/.data"
spark_session_service = SparkSessionService()
spark = spark_session_service.get_spark_session()
```

```python
from covid.common import read_delta

silver_path = f"{database_path}/silver/covid"
df = read_delta(spark, silver_path)
```

```python
from covid.gold import cases_per_region

cases_per_region_df = cases_per_region(df)
```

```python
cases_per_region_df.show(5)
```

```python
from covid.common import write_delta

gold_path = f"{database_path}/gold/covid"
write_delta(cases_per_region_df, gold_path, "overwrite")
```

```python
from covid.common import read_delta

df = read_delta(spark, gold_path)
```

```python
df.show(5)
```
