```python
from covid.common import SparkSessionService
```

```python
workspace_path = "COVID-DATABRICKS"
database_path = f"/{workspace_path}/.data"
spark_session_service = SparkSessionService()
spark = spark_session_service.get_spark_session()
```

```python
spark.read()
```

```python
df = (
    spark.read.format("csv")
    .option("header", True)
    .option("inferSchema", True)
    .option("delimiter", ";")
    .load(f"{database_path}/raw/HIST_PAINEL_COVIDBR_2020_Parte1_05set2025.csv")
)
```

```python
df.write.format("delta").mode("overwrite").save(f"{database_path}/bronze/raw")
```
